```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%pylab inline # adjust the size of figures
pylab.rcParams['figure.figsize'] = (10, 6) # adjust the size of figures
```

```python
x = np.linspace(-10., 10., 50)
```

## 1. Plot a Sigmoid Function and its Derivative

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def deriv_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].spines['left'].set_position('center')
ax[0].spines['bottom'].set_position('center')
ax[0].spines['right'].set_position('center')
ax[0].spines['top'].set_position('center')
ax[0].plot(x, sigmoid(x))

ax[1].spines['left'].set_position('center')
ax[1].spines['right'].set_position('center')
ax[1].spines['top'].set_color('none')
ax[1].xaxis.set_ticks_position('bottom')
ax[1].plot(x, deriv_sigmoid(x))
```

## 2. Plot a Hyperbolic Tangent Function and its Derivative

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
def deriv_tanh(x):
    return 1 - np.square(tanh(x))
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].spines['left'].set_position('center')
ax[0].spines['bottom'].set_position('center')
ax[0].spines['right'].set_position('center')
ax[0].spines['top'].set_position('center')
ax[0].plot(x, tanh(x))

ax[1].spines['left'].set_position('center')
ax[1].spines['right'].set_position('center')
ax[1].spines['top'].set_color('none')
ax[1].xaxis.set_ticks_position('bottom')
ax[1].plot(x, deriv_tanh(x))
```

## 3. Plot a ReLU Function and its Derivative

```python
def ReLU(x):
    return np.where(x > 0, x, 0)

def deriv_ReLU(x):
    return np.where(x > 0, 1, 0)
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].spines['left'].set_position('center')
ax[0].spines['right'].set_position('center')
ax[0].spines['top'].set_color('none')
ax[0].xaxis.set_ticks_position('bottom')
ax[0].plot(x, ReLU(x))

ax[1].spines['left'].set_position('center')
ax[1].spines['right'].set_position('center')
ax[1].spines['top'].set_color('none')
ax[1].xaxis.set_ticks_position('bottom')
ax[1].plot(x, deriv_ReLU(x))
```

## 4. Plot a Leaky ReLU Function and its Derivative

```python
def leaky_ReLU(x):
    return np.where(x > 0.01 * x, x, 0.01 * x)

def deriv_leaky_ReLU(x):
    return np.where(x > 0.01 * x, 1, 0.01)
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].spines['left'].set_position('center')
ax[0].spines['right'].set_position('center')
ax[0].spines['top'].set_color('none')
ax[0].xaxis.set_ticks_position('bottom')
ax[0].plot(x, leaky_ReLU(x))

ax[1].spines['left'].set_position('center')
ax[1].spines['right'].set_position('center')
ax[1].spines['top'].set_color('none')
ax[1].xaxis.set_ticks_position('bottom')
ax[1].plot(x, deriv_leaky_ReLU(x))
```

## 5. Plot a Exponential Linear Unit Function and its Derivative

```python
def ELU(x, alpha):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def deriv_ELU(x, alpha):
    return np.where(x > 0, 1, ELU(x, alpha) + alpha)
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].spines['left'].set_position('center')
ax[0].spines['right'].set_position('center')
ax[0].spines['top'].set_color('none')
ax[0].xaxis.set_ticks_position('bottom')
ax[0].axis([-10, 10, -0.1, 5])
ax[0].plot(x, ELU(x, 0.05))

ax[1].spines['left'].set_position('center')
ax[1].spines['right'].set_position('center')
ax[1].spines['top'].set_color('none')
ax[1].xaxis.set_ticks_position('bottom')
ax[1].plot(x, deriv_ELU(x, 0.05))
```
